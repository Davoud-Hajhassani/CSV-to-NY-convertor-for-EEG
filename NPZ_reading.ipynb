## `npz` reading in python 

```python
import numpy as np, pandas as pd
#from pandas import DataFrame
```

```python
#read the `npz` file in python

test= np.load('dataset/bi2015a/subject_01_session_01.npz')

print(test.files) 
DATA=test['data']
STIM=test['stim']
print(DATA.shape , STIM.size) # DATA.shape: Number of Samples x Number of channels (except ground)
                              # STIM.shape: a Vector with the size of Number of Samples
print(pd.DataFrame(DATA)) 

print(sum(STIM))              #which should be the sum of Trigger and Target columns
number_of_NonTarget=np.where(STIM==1)[0].size #number of NonTarget
print("number_of_NonTarget:", number_of_NonTarget)
number_of_Target=np.where(STIM==2)[0].size #number of Target
print("number_of_Target:", number_of_Target)
```
